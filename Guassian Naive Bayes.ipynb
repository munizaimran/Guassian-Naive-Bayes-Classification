```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
columns =  ["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age","Outcome"]
df = pd.read_csv('data.txt', sep = ',', header = None, names = columns)
df
```

```python
X = df.iloc[:, 0: 8]
X
```

```python
y = df.iloc[:, -1]
y
```

```python
#splitting the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)

#fit() method
model = GaussianNB().fit(X_train, y_train)

#making predictions
predicted_y = model.predict(X_test)
predicted_y
```

```python
#checking accuracy
accuracy_score = accuracy_score(y_test, predicted_y) 
print (accuracy_score)
```

```python
#creating a sample dataframe to test our model

person = pd.DataFrame()
person['Pregnancies'] = [7]
person['Glucose'] = [60]
person['BloodPressure'] = [86]
person['SkinThickness'] = [34]
person['Insulin'] = [0]
person['BMI'] = [33.5]
person['DiabetesPedigreeFunction'] = [0.264]
person['Age'] = [50]

person
```
